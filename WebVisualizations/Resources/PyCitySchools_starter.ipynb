```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# define variables that will populate the snapshot dataframe
school_sum = len(school_data_complete["school_name"].unique())
student_sum = school_data_complete["student_name"].count()
avg_math = school_data_complete["math_score"].mean()
avg_reading = school_data_complete["reading_score"].mean()

# refer to the unmerged school_data set for list of school budgets
total_budget = school_data["budget"].sum()

# filter for math scores 70 or greater
passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70].count()["Student ID"]
pct_passing_math = passing_math / float(student_sum)*100

# filter for reading scores 70 or greater
passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70].count()["Student ID"]
pct_passing_reading = passing_reading / float(student_sum)*100

# filter for math AND reading scores 70 or greater
passing_overall = len(school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)])/student_sum*100

# Creating a summary DataFrame using above values
district_summary_df = pd.DataFrame({"Total Schools": [school_sum],
                                   "Total Students": [student_sum],
                                   "Total Budget": [total_budget],
                                   "Average Math Score": [avg_math],
                                   "Average Reading Score": [avg_reading],
                                   "% Passing Math": [pct_passing_math],
                                   "% Passing Reading": [pct_passing_reading],
                                   "% Passing Overall": [passing_overall]})
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Groupby school_name
grouped_school_df = school_data_complete.groupby("school_name")

# define variables that will populate the summary dataframe
# school_name = grouped_school_df["school_name"].count()
school_type = grouped_school_df["type"].unique()
school_student_count = grouped_school_df["Student ID"].count()
school_budget = grouped_school_df["budget"].unique()
per_student_budget = school_budget / school_student_count
school_avg_math = grouped_school_df["math_score"].mean().round(2)
school_avg_reading = grouped_school_df["reading_score"].mean().round(2)

# filter for math scores 70 or greater
school_passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
grouped_passed_math = school_passing_math.groupby("school_name")
passed_math_count = grouped_passed_math['Student ID'].count()
school_pct_pass_math = (passed_math_count / school_student_count * 100.0).round(2)

# filter for reading scores 70 or greater
school_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
grouped_passed_reading = school_passing_reading.groupby("school_name")
passed_reading_count = grouped_passed_reading['Student ID'].count()
school_pct_pass_reading = (passed_reading_count / school_student_count * 100.0).round(2)

# filter for math AND reading scores 70 or greater
# school_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
# grouped_passed_reading = school_passing_reading.groupby("school_name")
# passed_reading_count = grouped_passed_reading['Student ID'].count()
# school_passing_overall = len(school_data_complete[school_data_complete["math_score"] >= 70] & school_data_complete[school_data_complete["reading_score"] >= 70])/school_student_count*100
school_passing_overall = ((school_pct_pass_math + school_pct_pass_reading) / 2).round(2)

# Create overview table to summarize all the metrics
school_summary_df = pd.DataFrame({"School Type": school_type,
                                  "Total Students": school_student_count,
                                  "Total School Budget": school_budget,
                                  "Per Student Budget": per_student_budget,
                                  "Average Math Score": school_avg_math,
                                  "Average Reading Score": school_avg_reading,
                                  "% Passing Math": school_pct_pass_math,
                                  "% Passing Reading": school_pct_pass_reading,
                                  "% Passing Overall": school_passing_overall
                        })

school_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# groupby school name, create a df with average math score (9,10,11,12) as columns
math_grade9_df = school_data_complete[(school_data_complete["grade"]=="9th")]
math_grade10_df = school_data_complete[(school_data_complete["grade"]=="10th")]
math_grade11_df = school_data_complete[(school_data_complete["grade"]=="11th")]
math_grade12_df = school_data_complete[(school_data_complete["grade"]=="12th")]

# groupby ("school name") w/ .mean()
grouped_math_grade9_df = math_grade9_df.groupby("school_name").mean()
grouped_math_grade10_df = math_grade10_df.groupby("school_name").mean()
grouped_math_grade11_df = math_grade11_df.groupby("school_name").mean()
grouped_math_grade12_df = math_grade12_df.groupby("school_name").mean()

# create final df with columns (9,10,11,12)
math_by_grade_df = pd.DataFrame({"9th": grouped_math_grade9_df["math_score"],
                                "10th": grouped_math_grade10_df["math_score"],
                                "11th": grouped_math_grade11_df["math_score"],
                                "12th": grouped_math_grade12_df["math_score"]
                                }).round(2)
math_by_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# columns: avg math score, avg reading score, % pass math, % pass reading, % overall pass rate
# rows: dependent bins ... small med large ... spending ranges

# Create bins in which to place values based upon average per student spending
bins = [550, 600, 650, 700]
        
# Create labels for these bins
group_labels = ["550 to 599", "600 to 649", "650 to 700"]
                
# Slice the data and place it into bins
pd.cut(school_summary_df["Per Student Budget"], bins, labels=group_labels).head()

# Place the data series into new columns inside of the DataFrame
school_summary_df["Spend Group"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels=group_labels)

# Create a GroupBy object based upon "Spend Group"
spending_group = school_summary_df.groupby("Spend Group")

# Find how many rows fall into each bin
# print(spending_group["Total Students"].count())

# Get the average of each column within the GroupBy object
spending_group[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Passing Overall"]].mean().round(2)
```
